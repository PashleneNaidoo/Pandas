# Series
  - A Series is a one-dimensional array. Also known as a column.


```python
# We create a list, in this instance we create a list of cars.
cars = ['Ferarri','Mercedes','Royals Royce','Porsche']

# We then feed the 'cars' list into the pd.Series() function to create a series 'mycars'
mycars = pd.Series(cars)

# Run the series
mycars
```

```python
# We create another list describing the number of cars of each kind.
quantity_of_cars = [2,10,6,1]

# Feed the list into the pd.Series() function to create a series 'myquantity_cars'
myquantity_cars = pd.Series(quantity_of_cars)

# Run the series
myquantity_cars
```

# DataFrame
  - A dataframe is 2 dimensional array. We can think of table with columns and rows.

```python
# We can combine two series to create a DataFrame.
# We use the pd.DataFrame() function and feed the two series with their column names
# Seperated by a "," comma . To create a Dataframe 'garage_home'.

garage_home = pd.DataFrame({
    'Car Brand': mycars,
    'Number of Cars': myquantity_cars
})

# Run the DataFrame
garage_home
```

```python
 # A Dataframe can be create using a list of tuples.
 garage_cpt = [('Lamboghini',1),('Mustang',4),('BMW',1),('Porsche',3)]

 #The list is fed into the dataframe function and
 # the columns are defined respectively to the tuple.
 mygarage_cpt = pd.DataFrame(garage_cpt,columns=['Car Brand','Number of Cars'])

 # Run the dataframe
 mygarage_cpt
```

```python
# Create a three column dataframe with three inputs in each tuple.
garage_monaco = [('Ferarri',1,'red'),('Mercedes',4,'white'),('Lamboghini',1,'purple'),('Royals Royce',1,'black'),('Porsche',3,'blue')]

#The list is fed into the dataframe function and
# the columns are defined respectively to the tuple.
mygarage_monaco = pd.DataFrame(garage_monaco,columns=['Car Brand','Number of Cars','Color of Car'])

# Run the dataframe
mygarage_monaco
```

# Cleaning Data


```python
# Let begin cleaning our Data.
mydf = df.set_index('ISRC')
```

```python
# The indices have been replaced by 'ISRC'
mydf.head()
```

```python
# To locate a specific song we need to its unique ISRC code.
mydf.loc['USSM12209777'] # previously at index 3
```

```python
# Describes the shape of our data
mydf.shape
```

```python
# Lets create a dataframe with data that we want to work with.
my_songs = mydf[['Track', 'Artist', 'Release Date',
         'All Time Rank', 'Spotify Streams', 'YouTube Views',
         'YouTube Likes','TikTok Views','TikTok Likes']]

# Run the dataframe
my_songs
```

```python
# Describes the data types in the my_songs dataframe.
my_songs.dtypes
```

# Changing DataTypes

```python
# Lets change the 'All Time Rank' column to integer type.
#my_songs['All Time Rank']

#step 1 : Look at data and identify what it contains. It has a bunch of commas.
# It should contain a point or just a number.
#my_songs['All Time Rank'].astype(str).str.replace(',', '')

#step2 : The data contains only numbers. We change the data type to an integer.
#my_songs['All Time Rank'] = my_songs['All Time Rank'].astype(str).str.replace(',', '').astype('int')

```

```python
# Convert 'All Time Rank' column to string type before using .str accessor
pd.options.mode.copy_on_write = True
mydf['All Time Rank'] = mydf['All Time Rank'].astype(str).str.replace(',', '').astype('int') # Remove commas before converting to integer
```

```python
# Convert 'Spotify Streams' column to string type, remove commas, and then convert to integer
my_songs['Spotify Streams'] = my_songs['Spotify Streams'].fillna('0').astype(str).str.replace(',', '').astype('int') # Explicitly convert to string type after filling NaN
```

```python
# Convert 'YouTube Views' column to string type, remove commas, fill NaN with '0', and then convert to integer
my_songs['YouTube Views'] = my_songs['YouTube Views'].str.replace(',', '').fillna('0').astype('int') # Fill NaN with '0' before conversion
```

```python
# Convert 'YouTube Likes' column to string type, remove commas, fill NaN with '0', and then convert to integer
my_songs['YouTube Likes'] = my_songs['YouTube Likes'].str.replace(',', '').fillna('0').astype(int) # Clean and convert
```

```python
#TikTok Views
my_songs['TikTok Views'] = my_songs['TikTok Views'].str.replace(',', '').fillna('0').astype(int) # Clean and convert
```

```python
#TikTok Likes
my_songs['TikTok Likes'] = my_songs['TikTok Likes'].str.replace(',', '').fillna('0').astype(int) # Clean and convert
```

```python
my_songs['Release Date'] = pd.to_datetime(my_songs['Release Date']) # Convert 'Release Date' column to datetime
```

```python
# The my_songs dataframe contains the correct data types.
my_songs.dtypes
```

```python
# Cleaned Data
my_songs.head()
```

# Using the Cleaned Data

```python
# Lets use our cleaned data.
# Lets find the top 10 songs currently in 2024.
# The .loc[] function locates the data in the 'All Time Rank' column that is less than 11.
my_TopTen_songs = my_songs.loc[mydf['All Time Rank'] < 11]

# Run the dataframe
# Top ten songs in 2024 currently on spotify.
my_TopTen_songs
```

# Creating a new column

```python
# A new column can be created using existing columns.
# We create a 'Total Views' column by adding the youtube views and tiktko view columns.
my_songs['Total Views'] = my_songs['YouTube Views'] + my_songs['TikTok Views']

# Run the dataframe
my_songs.head()
```
